### 状态模式把每个状态做一个类，再加一个管理类（环境），每个状态管理自己的行为及根据条件改变状态

```python
###例子  网络管理类
class StateBase(object):
    def __init__(self,net):
        self._net = net
    @property
    def net(self):
        return self._net
    def run(self):
        pass
#class Connected
class Connect(StateBase):
    def __init__(self,net):
        StateBase.__init__(self,net)
    def run(self):
        print('状态变为运行')
        self.net.setState(Connected(self.net))
class Connected(StateBase):
    def __init__(self,net):
        StateBase.__init__(self,net)
    def run(self):
        print('发送数据')
        print('发送数据')
        print('发送数据')
        print('发送数据')
        print('发送数据错误')
        self.net.setState(Connect(self.net))
class Net(object):
    def __init__(self):
        self._sendList = []
        self._reciveList = []
        self.setState(Connect(self))
    def send(self,msg):
        self._sendList.append(msg)
    def get(self):
        if len(self._reciveList) > 0:
            self._reciveList.pop()
        return None
    def setState(self,state):
        self._state = state
        self._state.run()
```
