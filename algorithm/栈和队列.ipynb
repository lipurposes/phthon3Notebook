### 栈
    栈是后进先出的线性表，只操作栈顶数据，一般用数组的方式实现

```python
#栈的数组实现
class Stack(object):
    def __init__(self):
        self.__data = []
    def pop(self):
        return self.__data.pop()
    def stackValue(self):
        if len(self.__data) > 0:
            return self.__data[len(self.__data) - 1]
        return None
    def push(self,value):
        self.__data.append(value)
    def __len__(self):
        return len(self.__data)
    def __str__(self):
        return " ".join('%s' %v for v in self.__data)
#栈的测试
tStack = Stack()
tStack.push(24)
tStack.push(15.2)
tStack.push('arrr')
print(tStack)
print('删除一个元素：')
tStack.pop()
print('栈长度为：',len(tStack))
print('栈最终为：',tStack)
```

### 队列
    队列是先进先出的线性表，一般实现方式为链表

```python
#队列的链表实现，python模拟
class Node(object):
    def __init__(self,value):
        self.__value = value
        self.__next = None
    @property
    def next(self):
        return self.__next
    @next.setter
    def next(self,nextP):
        self.__next = nextP
    @property
    def value(self):
        return self.__value
class myQueue(object):
    def __init__(self):
        self.__front = Node(None)
        self.__rear = self.__front
    def push(self,node):
        self.__rear.next = node
        self.__rear = node
    def front(self):
        retNode = self.__front.next
        if self.__front.next == self.__rear :
            self.__rear = self.__front
        if self.__front.next != None :
            self.__front.next = self.__front.next.next
        return retNode
    def __str__(self):
        retStr = ''
        showN = self.__front.next
        while showN != None:
            retStr += str(showN.value)
            retStr += ' '
            showN = showN.next
        return retStr
#测试
testQ = myQueue()
testQ.push(Node(53))
testQ.push(Node('str1'))
testQ.push(Node('str2'))
print('当前的队列为',testQ)
testQ1 = testQ.front()
print('队首为：',testQ1.value)
print('当前队列为：',testQ)
testQ1 = testQ.front()
print('队首为：',testQ1.value)
print('当前队列为：',testQ)
testQ1 = testQ.front()
print('队首为：',testQ1.value)
print('当前队列为：',testQ)
```
