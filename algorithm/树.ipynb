    1.孩子兄弟表示法是将多子树转化为二叉树的方法，转化后的二叉树根节点只有左子树
    2.将森林转化为二叉树方法，每一颗树通过孩子兄弟表示法转化为二叉树，以第一棵树的根节点作为根节点，其余的树作为根节点的兄弟节点连接为二叉树

    2.完全树（满树失去后边一些节点的树） 可以用堆表示

```python
class TreeNode(object):
    def __init__(self,data):
        self.__data = data
        self.__left = None
        self.__right = None
    @property
    def left(self):
        return self.__left
    @left.setter
    def left(self,value):
        self.__left = value
    def data(self):
        return self.__data
    @property
    def right(self):
        return self.__right
    @right.setter
    def right(self,value):
        self.__right = value
def setTreeNode(treeStr,pos):
    print(treeStr,pos)
    if pos > len(treeStr) or treeStr[pos] == '^':
        return None,pos + 1
    treeN = TreeNode(treeStr[pos])
    pos += 1
    treeN.left,pos = setTreeNode(treeStr,pos)
    treeN.right,pos = setTreeNode(treeStr,pos)
    return treeN,pos
def getTreeNode(treeNode,level):  ## 前序遍历
    if treeNode :
        print('第 %d 层 ：%s' %(level,treeNode.data()))
        getTreeNode(treeNode.left,level + 1)
        getTreeNode(treeNode.right,level + 1)
##前序遍历的非递归实现(利用栈)
def getTreeNode1(treeNode):
    stackList = []
    while treeNode :
        print('%s' % treeNode.data())
        if treeNode.right :
            stackList.append(treeNode.right)
        if treeNode.left :
            treeNode = treeNode.left
        elif len(stackList) > 0 :
            treeNode = stackList.pop(0)
        else :
            treeNode = None
    
```

### 赫夫曼编码
    利用二叉树构建数据中字符（字节）的权重树，根据字符出现的频率用变字节编码，从而压缩数据，编码为前缀码（所有字符的编码不是另一字符的前缀）
    1.首先生成每个字符的频率表
    2.按照频率从高到低排序频率表
    3.生成最优二叉树
    4.通过最优二叉树生成字符编码（每往左子树走一层，加0，每往右子树走一层，加1）

```python
##生成字符的频率表
def getByteCounts(byteValue):
    btArray = bytearray(byteValue)
    byteMap = {}
    for bt in btArray:
        if byteMap.get(bt):
            byteMap[bt] += 1
        byteMap[bt] = 1
    return byteMap
##测试
lStr = "我们的明天"
lt = getByteCounts(lStr.encode('utf-8'))
print(lt)                    
```

```python
class ListNode(object):
    def __init__(self):
        self._leftChild = None
        self._rightChild = None
        self._value = None
    @property
    def value(self):
        return self._value
    @value.setter
    def value(self,sp):
        self._value = sp
    @property
    def leftChild(self):
        return self._leftChild
    @leftChild.setter
    def leftChild(self,child):
        self._leftChild = child
    @property
    def rightChild(self):
        return self._rightChild
    @rightChild.setter
    def rightChild(self,child):
        self._rightChild = child
class ListItem(object):
    def __init__(self):
        self._node = ListNode()
        self._next = None
    @property
    def weight(self):
        return self._count
    @weight.setter
    def weight(self,num):
        self._count = num
    @property
    def node(self):
        return self._node
    @property
    def next(self):
        return self._next
    @next.setter
    def next(self,nexter):
        self._next = nexter
class Lists(object):
    def __init__(self):
        self._count = 0
        self._first = None
    def addItem(self,item):
        if self._count == 0:
            self._first = item
        elif item.weight <= self._first.weight:
            item.next = self._first
            self._first = item
        else :
            cur = self._first
            while cur.next :
                if item.weight <= cur.next.weight :
                    item.next = cur.next
                    cur.next = item
                    break
                cur = cur.next
            if not cur.next :
                cur.next = item
        self._count += 1
    def getItem(self):
        if self._count > 0 :
            self._count -= 1
            retItem = self._first
            self._first = self._first.next
            return retItem
        return None
    @property
    def first(self):
        return self._first
    @property
    def count(self):
        return self._count
    
def createCode(codeTable,listNode,strList,strP):
    if listNode.value :
        useStrList = strList[0:strP]
        codeTable[listNode.value] = ''.join(useStrList)
    else:
        if listNode.leftChild:
            strList[strP] = '0'
            createCode(codeTable,listNode.leftChild,strList,strP + 1)
        if listNode.rightChild:
            strList[strP] = '1'
            createCode(codeTable,listNode.rightChild,strList,strP + 1)
# 获取编码表    
def getCodeTable(listNode):
    codeTable = {}
    strList = [0] * 50
    strP = 0
    createCode(codeTable,listNode,strList,strP)
    return codeTable
# 用二进制编码的次数生成从小到大排列的链表
def getLinkedList(countDict):
    weightList = Lists()
    for key,value in countDict.items():
        listItem = ListItem()
        listItem.weight = value
        node = listItem.node
        node.value = key
        weightList.addItem(listItem)
    return weightList

#获取最优二叉树
def getBiTree(linkedList): 
    while linkedList.count > 1 :
        print(linkedList.count,end= ' ')
        left = linkedList.getItem()
        right = linkedList.getItem()
        listItem = ListItem()
        listItem.weight = left.weight + right.weight
        listItem.node.leftChild = left.node
        listItem.node.rightChild = right.node
        linkedList.addItem(listItem)
    return linkedList.first.node
#根据转换后的编码字符串生成原字符串
def getPrimeStr(strToCodeDict,codeStr):
    sB = bytearray()
    strLen = len(codeStr)
    pF = 0
    pL = 1
    while pL <= strLen:
        testStr = codeStr[pF:pL]
        if strToCodeTable.get(testStr):
            sB.append(strToCodeTable[testStr])
            pF = pL
            pL = pF + 1
        else:
            pL += 1
    return sB.decode('utf-8')
#将01字符串转换为对应byte
def strToByte(codeStr):
    btArray = bytearray()
    pos = 0
    strLen = len(codeStr)
    while pos < strLen:
        byteStr = ''
        if pos + 8 > strLen:
            byteStr = codeStr[pos:]
        else:
            byteStr = codeStr[pos:pos + 8]
        print('当前byte',byteStr)
        intV = int(byteStr,2)
        btArray.append(intV)
        pos = pos + 8
    return bytes(btArray)
#将01byte转换为01字符串
#这儿有个问题，最后的byte应该转换为几位str？所以这里加了个总位数字段
def byteToByteStr(byteC,lenCode):
    byteStr = ''
    for i in range(len(byteC)):
        bStr = bin(byteC[i]).replace('0b','')
        if i == len(byteC) - 1:
            prmL = lenCode - 8 * (len(byteC) - 1)
            bStr = ''.join(['0'] * (prmL - len(bStr))) + bStr
        else:
            bStr = ''.join(['0'] * (8 - len(bStr))) + bStr
        byteStr += bStr
    return byteStr
```
